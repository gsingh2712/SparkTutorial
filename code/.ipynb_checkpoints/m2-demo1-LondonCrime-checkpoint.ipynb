```python
## Creating SparkSession Entry point to Spark Application
from pyspark.sql import SparkSession
```

```python
## Creating Spark Session using builder
spark = SparkSession.builder\
                    .appName("Analyzing London Crime Data ")\
                    .getOrCreate()
```

```python
data = spark.read\
            .format("csv")\
            .option("header", "true")\ ## passing so that headers are interpreeted correctly
            .load("../datasets/london_crime_by_lsoa.csv")
```

```python
data.printSchema()
```

```python
## see a subset 
data.limit(5).show()
```

# DROP Rows that have NA's -> values not available

```python
data.dropna()
```

```python
## Trimming Data Not needed so that we can work faster
data = data.drop('lsoa_code')
data.show()
```

```python
## Getting distinct values
total_boroughs = data.select("borough")\
                        .distinct()
total_boroughs.show()
```

```python
total_boroughs.count()
```

## Filtering

```python
hackney_data = data.filter(data['borough'] == "Hackney")
hackney_data.show()
```

```python
## Check use of isin
data_2015_2016= data.filter(data['year'].isin(["2015", "2016"]))
data_2015_2016.sample(fraction=0.1).show()
```

```python
data_2014_onwards = data.filter(data['year'] >= 2014)
data_2014_onwards.sample(fraction=0.1).show()
```

## Aggregation on DF

```python
## Group By
borough_crime_count = data.groupBy("borough")\
                            .count()
borough_crime_count.show(5)
```

```python
borough_conviction_sum = data.groupBy("borough")\
                                .agg({"value":"sum"})
borough_conviction_sum.show(5)
```

```python
borough_conviction_sum = data.groupBy("borough")\
                                .agg({"value":"sum"})\
                                .withColumnRenamed("sum(value)", "convictions")
borough_conviction_sum.show(5)
```

```python
## Total Convictions
total_borough_convictions = data.agg({"value":"sum"})
total_convictions.show()
```

```python
total_convictions = total_borough_convictions.collect()[0][0]
```

```python
import pyspark.sql.functions as func
```

```python
### Adding Collumn on Fly
borough_percentage_conviction = borough_conviction_sum.withColumn(
    "% contribution",
    func.round(borough_conviction_sum.convictions / total_convictions * 100 ,2))
borough_percentage_conviction.printSchema()
```

```python
### Order By % contribution column in descending order
borough_percentage_conviction.orderBy(borough_percentage_conviction[2].desc())\
    .show(10)
```

```python
conviction_monthly = data.filter(data["year"]=="2014")\
                        .groupBy("month")\
                        .agg({"value":"sum"})\
                        .withColumnRenamed("sum(value)", "convictions")
```

```python
conviction_monthly.show()
```

# CROSSTAB FUNCTION VERY IMPORTANT

## Here we want to see how borough varies across categories of crimes

```python
data.crosstab('borough','major_category')\ ## only for four crimes
            .select('borough_major_category','Burglary' , 'Drugs' , 'Fraud or Forgery' , 'Robbery' )\
            .show()
```

```python
get_ipython().magic('matplotlib inline')
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def describe_year(year):
    yearly_details = data.filter(data.year == year)\
                        .groupBy('borough')\
                        .agg({'value':'sum'})\
                        .withColumnRenamed("sum(value)", "convictions")
    
    borough_list = [ x[0] for x in yearly_details.toLocalIterator() ]
    convictions_list = [ x[1] for x in yearly_details.toLocalIterator() ]
    plt.figure(figsize=(33, 10))
    plt.bar(borough_list , convictions_list)
    
    plt.title('Crime for the year: ' + year , fontsize=30)
    plt.xlabel('Boroughs', fontsize=30)
    plt.ylabel('Convictions', fontsize=30)
    
    plt.xticks(rotation=90 , fontsize=30)
    plt.yticks(fontsize=30)
    plt.autoscale()
    plt.show()
```

```python
describe_year('2014')
```
